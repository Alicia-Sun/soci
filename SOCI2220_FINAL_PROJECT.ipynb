```python
import pandas as pd

file_path = '/content/Share_of_green_areas_and_green_area_per_capita_in_cities_and_urban_areas_1990_-_2020.xlsx'
data = pd.read_excel(file_path, sheet_name='Data')
data.head()
```

```python
us_data = data[data['Country or Territory Name'] == 'United States of America']
us_data.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = us_data
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# Histogram of green area values for 2020
plt.figure(figsize=(10, 6))
sns.histplot(df['Green area per capita 2020 (m2/person)'], bins=20, kde=True)
plt.title("Distribution of Green Area per Capita (2020)")
plt.xlabel("Green Area per Capita (m2/person)")
plt.ylabel("Frequency")
plt.show()
```

```python
top_cities = df[['City Name', 'Green area per capita 2020 (m2/person)']].sort_values(by='Green area per capita 2020 (m2/person)', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Green area per capita 2020 (m2/person)', y='City Name', data=top_cities, palette="viridis")
plt.title("Top 10 Cities by Green Area per Capita (2020)")
plt.xlabel("Green Area per Capita (m2/person)")
plt.ylabel("City Name")
plt.show()
```

```python
import statsmodels.api as sm

df_analysis = df[['# people at risk for severe depression per 100k of county population', 'Average share of green area in city/ urban area 2020 (%)']]

correlation = df_analysis.corr().loc['# people at risk for severe depression per 100k of county population', 'Average share of green area in city/ urban area 2020 (%)']
print(f"Correlation between Depression Risk and Green Area: {correlation:.2f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Average share of green area in city/ urban area 2020 (%)', y='# people at risk for severe depression per 100k of county population', data=df_analysis, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title("Relationship Between Green Area and Depression Risk")
plt.xlabel("Green Area per Capita (m2/person)")
plt.ylabel("People at Risk for Severe Depression (per 100k)")
plt.show()

# Simple Linear Regression
X = df_analysis['Average share of green area in city/ urban area 2020 (%)']
y = df_analysis['# people at risk for severe depression per 100k of county population']
X = sm.add_constant(X)  # Add a constant for the regression model

model = sm.OLS(y, X).fit()
print("\nRegression Results:")
print(model.summary())
```

```python
import statsmodels.api as sm

df_analysis = df[['# people at risk for severe depression per 100k of county population', 'Green area per capita 2020 (m2/person)']]

correlation = df_analysis.corr().loc['# people at risk for severe depression per 100k of county population', 'Green area per capita 2020 (m2/person)']
print(f"Correlation between Depression Risk and Green Area: {correlation:.2f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Green area per capita 2020 (m2/person)', y='# people at risk for severe depression per 100k of county population', data=df_analysis, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title("Relationship Between Green Area and Depression Risk")
plt.xlabel("Green Area per Capita (sq. meters)")
plt.ylabel("People at Risk for Severe Depression (per 100k)")
plt.show()

# Simple Linear Regression
X = df_analysis['Green area per capita 2020 (m2/person)']
y = df_analysis['# people at risk for severe depression per 100k of county population']
X = sm.add_constant(X)  # Add a constant for the regression model

model = sm.OLS(y, X).fit()
print("\nRegression Results:")
print(model.summary())
```

This suggests that green area per capita has minimal explanatory power for predicting depression risk. There is no statistically significant relationship between green area per capita and the number of people at risk for severe depression in this dataset. Green area per capita explains only a very small proportion of the variability in depression risk.

```python
time_columns = ['Green area per capita 1990 (m2/person)', 'Green area per capita 2000 (m2/person)', 'Green area per capita 2010 (m2/person)', 'Green area per capita 2020 (m2/person)']
df_time_trends = df[time_columns].mean()
renamed_columns = ['1990', '2000', '2010', '2020']
df_time_trends.index = renamed_columns

plt.figure(figsize=(10, 6))
df_time_trends.plot(kind='line', marker='o')
plt.title("Trend of Green Area Per Capita Over Time")
plt.xlabel("Year")
plt.ylabel("Average Green Area per Capita (m2/person)")
plt.xticks(rotation=20)
plt.grid(True)
plt.show()
```
